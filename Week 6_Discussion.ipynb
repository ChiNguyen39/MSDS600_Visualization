```python
import pandas as pd
```

```python
df_path = '/Users/kie/Documents/Regis University/MS/01. MSDS 600/06. Week 6/Discussion/vehicles.csv'
df = pd.read_csv(df_path)
df
```

```python
df.isna().sum()
```

```python
import missingno as msno
import matplotlib.pyplot as plt
msno.matrix(df,filter='top',sort='ascending')
plt.show()
```

```python
df = df.drop('condition',axis=1)
df = df.drop('cylinders',axis=1)
df = df.drop('VIN',axis=1)
df = df.drop('drive',axis=1)
df = df.drop('size',axis=1)
df = df.drop('county',axis=1)
df
```

```python
df= df.dropna()
```

```python
msno.matrix(df,filter='top',sort='ascending')
plt.show()
```

```python
df['manufacturer'].value_counts().sort_values(ascending=False)[:10].plot.bar()
plt.title('Top manufacturers with top sold quantity')
```

```python
df.groupby('manufacturer').price.mean().sort_values(ascending=False)[:10].plot.bar()
plt.title('Top manufacturers having highest price vehicles')
```
